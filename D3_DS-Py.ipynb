#### Separate input and output variables

```python
x = df['YearsExperience'].values # input
y = df['Salary'] # output
```

```python
x
```

```python
y;
```

```python
x = x.reshape(30, 1) # 30 rows and 1 column
x
```

#### import the class

```python
from sklearn.linear_model import LinearRegression
```

#### create the object of the class (linear regression)

```python
regressor = LinearRegression()
```

#### Train the algorithm with data

```python
regressor.fit(x, y)
```

#### predict on unseen data

```python
# how much I will get after 15 years
```

```python
regressor.predict([[15]])
```

```python
regressor.predict([[20]]) # after 20 years
```

```python
regressor.predict([[0]]) # salary of a fresher
```
